# Large Language Models

Large Language Models are part of Foundation Models. These are mainly trained on massive amounts of text corpus, usually billions of parameters. These systems have the power to interact as humans. It all started with the famous paper from Google - Attention is All You Need! and Transformer models which entirely changed the AI landscape where models were better able to understand natural language and word meanings, correlations, and even come up with multimodal functionalities. 

There are open-source models as well as closed-source models. OpenAI, the GenAI giant has its patent GPT models, GPT3/4 generates text, translates languages, writes different kinds of creative content, and answers your questions in an informative way. HuggingFace BLOOM works well on generating text rather than other tasks but it can be made to do other things as well with proper prompt engineering. Recent ones are Googles PALM2, Meta AIs LLAMA and many more from Cohere and AI24Labs. These systems can be down streamed to perform text-specific tasks such as text generation (next word/phrase predictors) seen in auto-replies for teams, mails, code generation (GitHub copilot, AWS Code Whisperer), information retrieval (semantic search), etc.

- Once we have the base model from large unsupervised data, we can fine-tune it over a small supervised dataset. Parameter Efficient Fine-Tuning (PEFT) can help achieve such results. As models grow larger, fine-tuning becomes a challenging task given the hardware consumption dependencies. Fine-tuned models incur massive charges in storing and deploying as the fine-tuned model is of the same size as the original pre-trained model. Parameter-Efficient Fine-tuning (PEFT) is developed keeping in mind both of these bottlenecks. PEFT deals with only fine-tuning a small number of (extra) model parameters against freezing most model parameters of the pre-trained LLMs, which helps decrease the computational and storage costs. This approach has the additional benefit of catastrophic forgetting, a common behaviour observed in models during the full finetuning of LLMs.

- Reward modelling is when the loss function measures the predicted rewards consistency with the labelled ordering. In Reward model generation the ultimate goal is to produce a model or system that is capable of taking in a sequence of text and returns with a scalar reward that contains numeric representation. This leads us to the next step.

- Lastly, the best method is RLHF which surpasses all the avoid methods. It is a way to use reinforcement learning to optimise LLM with human feedback. It consists of 3 main stages - Pretraining a language model (LM), gathering data and training a reward model, and fine-tuning the LM with reinforcement learning. RLHF trains LLMs by learning directly from evaluations by humans about their performance. The prediction or decision LLMs take that is incorrect or suboptimal, the feedback from humans is used to correct the error or suggest alternatives. Over time, the model can self-learn and come up with better responses. RLHF is useful in complex tasks where it's unclear to define algorithmic solution but where humans judgement can enhance the quality of the LLM's output (e.g. if the task is to generate an essay, humans can score the LLM-generated outputs and the LLM uses the feedback to improve its generation).


# Prompt Engineering

Prompt-based ML send requests to LLMs describing the task that it is supposed to perform. These could be text-to-text or text-to-image/video/audio. Prompts help to interact with LLMs easier. The more precise the prompt the better the output. Both prompts and completions are part of the context length set by each LLM. On average it's 4k-8k depending on model to model. Following are the different types of prompting techniques.

* Zero-shot prompting: This is when a model can generate answers for situations it was not explicitly trained on. This doesnt require a retraining model on any data. Hence it's not aware of the features of the new task and in general comes up with answers which might not be relevant.

* Few-shot prompting: As the name suggests is based on training the model with few examples. This could be beneficial for tasks where data is not readily available. Hence it has limitations of learning entirely about the new task.

* Chain-of-Thoughts(CoT) Prompting - This process helps to understand the reasoning by providing additional information. Mathematical problems which need a step-wise explanation of how to arrive at the final answer. 

* Self-Consistency - This process lets the model run multiple times over the same prompt to generate different outputs. The majority of all the answers are taken as the final output.

* Tree-of-Thoughts - This process generates a tree with the decisions it takes to solve a problem. It allows LLM to self-evaluate the reasoning in making decisions and evaluate using breadth-first search and depth-first search and explore through backtracking.

* Fine-tuning LLMs is often helpful with domain-specific data using prompts & completions format. Fine-tuning is an old way where the top layers of the neural network get replaced with weights from new datasets.

# LangChain

LangChain is the LLM framework that can be used to interact with models, set up prompt templates, LLM Agents, chain replies and many more. Langchain has vast support of models from AI24Labs, Anthropic, Cohere, HuggingFace, OpenAI and several others. Langchain makes it easier for developers making it a one-stop for building LLM applications. As of now, it has support in Javascript & Python. Langchain consists of several modules : 

* Model Hub - Interface with language models
* Data connection - Interface with application-specific data
* Chains - Construct sequences of calls
* Agents - Let chains choose which tools to use given high-level directives
* Memory - Persist application state between runs of a chain
* Callbacks - Log and stream intermediate steps of any chain

Langchain has become popular with its vast support of API usage and a variety of plugins to put up an LLM application. Let us understand what is the gap that Langchain is filling. Suppose you have a banking client coming to you saying they need to build a chatbot which reads multiple banking-related documents and PDFs and answers questions from those documents and PDFs. You cannot go to ChatGPT every time as it would not have any previous context memory and also context length, and token length limitations. This is where Langchain steps in providing an LLM which can process this banking information from documents saved using vector database indexes to create chains or memory. Not to worry there is also support for document loaders and transforming into character chunks and embeddings. This entire process can be plugged into any pipeline or web app easily and made use of. # Use of Vector Databases, Embeddings

Vector stores are a type of data store that is optimized for storing and retrieving high-dimensional vectors. Document chunks are transformed into embeddings. LLMs generate high-dimensional vectors as outputs, and vector stores can be used to store these vectors in a way that is efficient and fast to access by creating indexes. Various implementation options available are  Milvus (by Zilliz), Pinecone, Weaviate, Annoy (by Spotify) and FAISS (by Meta). These can help caching the LLM memory or retrieve quickly domain-specific data from large files. 

Vector databases are different from traditional SQL & NoSQL databases in the aspect of the way they represent data storage and retrieval. VectorDB are not structured in nature, they tend to store data and index in mathematical vectors in high-dimension space. Also known as vectorization, that allows more efficiency in tasks like similarity searches and better handling of complex data types, such as images, audio, and text. Embedding is the process of converting these complex data into numeric representations of vectors or matrices. This form of data representation makes it easier for deep neural network models to understand and work with it. Embeddings provide several benefits, including dimensionality reduction, capturing semantic relationships, and facilitating similarity-based computations. They are widely used in tasks such as recommendation systems, text classification, information retrieval, and clustering.

The main advantage of using vector databases is how fast and accurate they are in similarity search and retrieval, based on vector distance. This works better than query-based databases which try to find exact matches, a vector database gets the most similar or relevant data based on semantic or contextual meaning. Below are usage strategies for using vector databases:

1. Efficient Storage and Retrieval: Vector databases are specifically designed to store and retrieve high-dimensional vector data efficiently. They employ specialized indexing and search algorithms, such as approximate nearest neighbor (ANN) search, to perform fast and accurate similarity-based queries. This makes them well-suited for applications that require similarity matching, such as recommendation systems or image/audio search.

2. Native Support for Vector Operations: Vector databases provide native support for vector operations, including distance calculations, vector arithmetic, and aggregations. This allows for complex computations directly within the database, eliminating the need for transferring data to external processing frameworks. It can significantly speed up analysis and reduce computational overhead.

3. Flexibility in Data Model: Vector databases offer flexibility in terms of data model and schema. They can handle unstructured or semi-structured data, such as text documents, images, or sensor readings, represented as vectors. This enables efficient storage and retrieval of diverse data types without the need for complex data transformations or normalization.

4. Scalability and Performance: Vector databases are designed to scale horizontally, allowing for distributed storage and parallel processing. This enables handling large volumes of data and high query loads efficiently. By distributing the data across multiple nodes, vector databases can achieve high throughput and low latency for real-time applications.

5. Machine Learning Integration: Vector databases can seamlessly integrate with machine learning workflows. They can store and manage vector embeddings or feature representations derived from machine learning models. This enables efficient retrieval and comparison of embeddings for tasks like similarity search, recommendation, or clustering.

6. Real-time Analytics: Vector databases can provide real-time analytics capabilities by performing complex operations on vector data at scale. They can handle streaming data, enabling immediate analysis and decision-making based on up-to-date information. This is particularly valuable in applications like fraud detection, anomaly detection, or real-time personalization.

7. Simplified Development and Deployment: Vector databases often provide user-friendly interfaces, APIs, and query languages tailored for vector-based operations. This simplifies development and integration into existing applications. Additionally, some vector databases offer cloud-based managed services, reducing the operational overhead of managing infrastructure.

Common vector database usage:

- Search images based on their visual content and style
- Search documents based on their topic and sentiment
- Search products based on their features and ratings



# Hallucinations & Grounding

It's common for large language models to hallucinate and provide answers which aren't correct but could seem to be correct. Since LLMs are trained on vast data nearly everything available on the internet, it often leads to confusion in reasoning and long context answers. For example, a hallucinating LLM when asked to predict the revenue generated by Tesla this year said it's $13.6, which is pretty much less than the actual. AI hallucinations are similar to human hallucinations but different in a way that AI makes it into interesting sentences which can make believe false information and unjustified answers, on the other hand, human hallucinations are based on false percepts. Researchers have identified hallucinations that could occur due to insufficient training data and adversarial hallucinations happening due to statistical phenomena. Following are potential causes of LLM hallucinations:

1. Overgeneralization: LLMs learn patterns from the training data, and sometimes they tend to overgeneralize these patterns. When faced with ambiguous or incomplete prompts, LLMs might generate responses that go beyond the intended meaning or extrapolate information that is not grounded in reality. This can lead to hallucinatory outputs.

2. Lack of Contextual Understanding: LLMs lack true understanding and context comprehension. They rely on statistical patterns in the data to generate responses. In some cases, they might misinterpret the context or fail to grasp the nuances of the prompt, resulting in hallucinations or incorrect information.

3. Biased Training Data: If the training data used for LLMs contains biased or misleading information, the model may reproduce and amplify those biases. Biases present in the data can influence the generated outputs, leading to hallucinatory or discriminatory responses.

4. Rare or Out-of-Distribution Inputs: LLMs are trained on vast amounts of text data, but they may encounter inputs that are rare or outside the distribution of the training data. When faced with such inputs, LLMs may struggle to generate appropriate responses and instead produce hallucinations or nonsensical outputs.

5. Unsupervised Pre-training: LLMs are often pre-trained in an unsupervised manner, where they learn from vast amounts of text data without specific task supervision. While this approach allows them to learn rich language representations, it can also result in the model generating imaginative or hallucinatory content based on the patterns it has learned.

6. Fine-tuning Bias: LLMs are typically fine-tuned on specific datasets or tasks, which can introduce biases or skewed perspectives. If the fine-tuning data is biased, it may contribute to the generation of hallucinations or outputs that align with those biases.

Grounding is a way to restrict models from hallucinating. Reinforcement learning and human feedback have proven to be the most useful methods in grounding LLMs. This is an ongoing research field and thus nothing in particular has come up which can mitigate hallucinations from models. Guardrails could be formed to stop LLMs from hallucinating with fact-checking from another LLM. Other techniques include:

1. Curate Diverse and Representative Training Data: Ensure that the training data used for LLMs is diverse, representative, and free from biases. Carefully curate the dataset to include a wide range of perspectives and avoid reinforcing stereotypes or misinformation. Consider incorporating data from multiple sources and perspectives to provide a balanced training signal.

2. Address Biases in Data: Implement techniques to detect and mitigate biases present in the training data. This can involve bias analysis, data preprocessing, or augmentation methods. Actively work on reducing both explicit and implicit biases to promote fair and unbiased generation.

3. Fine-Tuning with Adversarial Examples: During the fine-tuning phase, incorporate adversarial examples that challenge the LLM's behavior and encourage it to generate more accurate and reliable responses. Adversarial training can help the model identify and rectify hallucinatory tendencies.

4. Contextual Understanding and Reasoning: Enhance LLMs' contextual understanding and reasoning abilities. Incorporate external knowledge sources or knowledge graphs to provide structured information and improve reasoning capabilities. Encourage the model to reason over facts, logical relationships, and cause-and-effect patterns to produce more accurate and grounded responses.

5. Adversarial Testing and Evaluation: Develop robust evaluation methodologies that include adversarial testing. Evaluate the model's performance against challenging prompts or inputs designed to induce hallucinations or biases. This can help identify weaknesses and guide further improvements.

6. Transparent Model Explanations: Enhance transparency and explainability of LLMs. Develop methods to provide explanations for generated outputs, highlighting the model's reasoning process and the sources of information it relied upon. This can help users understand and identify potential hallucinations or biases.

7. Continuous Monitoring and Iterative Improvement: Establish ongoing monitoring mechanisms to detect and address hallucinations. Regularly evaluate the model's behavior in real-world scenarios and collect user feedback to identify and rectify issues. Continuously update and improve the model based on these findings.

8. Collaborative Efforts and Ethical Guidelines: Engage in collaborative efforts within the AI community to share best practices, research findings, and guidelines for responsible AI development. Encourage adherence to ethical guidelines and standards that prioritize fairness, transparency, and user well-being.

Grounding LLMs is an ongoing challenge, and it requires a combination of techniques from knowledge representation, natural language understanding, and information retrieval. By incorporating real-world knowledge, fact-checking mechanisms, and context awareness, we can improve the groundedness of LLMs and enhance their reliability in generating accurate and contextually appropriate responses.

# Bringing Reasoning to LLMs

Complex reasoning-based tasks are difficult for LLMs to handle and often lead to wrong answers or hallucinate. This is where SocraticAI steps in providing reasoning to LLMs with its novel methods - Multi-Agent Collaborative Problem Solving (with Humans in the Loop), free-form inquiry among agents to fully unlock the potential of LLMs, Rethinking Prompting for Knowledge and Reasoning.

It is built in ChatGPT 3.5. This framework mainly involves three LLM agents, who do a role-play, Socrates (an analyst who interacts with Theaetetus), Theaetetus (an analyst who interacts with Socrates), and Plato (a proofreader who mandates the decisions made) to solve problems collaboratively. The agents are given access to WolframAlpha and a Python code interpreter.

Key features of Socratic AI systems may include:

* Question Generation: Generate questions that prompt users to reflect, analyze, and explore different aspects of a topic. These questions are designed to encourage critical thinking and stimulate deeper understanding.

* Dialogue Management: Engage in interactive conversations with users, responding to their answers, probing further, and providing feedback or additional prompts. They follow a dialogue flow that facilitates a structured and guided exploration of ideas.

* Domain Knowledge: Incorporate domain-specific knowledge or access external knowledge sources to provide contextually relevant information during the questioning process. This helps users gain a deeper understanding of the topic under discussion.

* Adaptive Learning: Employ machine learning techniques to adapt their questioning strategies based on user responses, learning patterns of successful inquiry, and tailoring the experience to individual users' needs.

* Feedback and Evaluation: Provide feedback on user responses, offering insights into their thinking, identifying gaps in understanding, or suggesting alternative perspectives. They may also evaluate users' reasoning or solutions and provide guidance for improvement.

# Responsible Generative AI

Despite the advancements in generative AI, there are risks and harm involved. Every new invention has its good and bad sides and it's up to us which one we choose. The harms of AI can cost humanity in many ways, hence is high time generative AI be governed to mitigate bias, induce fairness and become responsible systems to serve the betterment of mankind. Many countries have started taking steps in regulating AI towards LLMs usage. AI systems should be ethical, transparent, explainable, reliable, sustainable, secure and safe to use for any individual or organisation. There are open-source tools available to ensure AI ethics which can help ensure proper policies and quality checks being following at every step of application development. Today almost every company has become AI-inclusive, which brings the question to be more and more cautious of what data is being used, what algorithms are incorporated, what evaluation metrics and standards are followed and finally how end-user acceptance comes through. Deepfakes are one the most dangerous aspects and have risen to better qualities increasing scams, and cybercrimes. Spread of misinformation spreads at a greater speed than that of genuine ones. 

1. Understand the Potential Risks: Begin by familiarizing yourself with the potential risks and challenges associated with generative AI systems. These may include biased or harmful outputs, data privacy concerns, or unintended consequences.

2. Establish Ethical Guidelines: Define a set of ethical guidelines that align with societal values and principles. These guidelines should inform the development and deployment of your generative AI system. Consider involving diverse stakeholders, including ethicists, domain experts, and affected communities, in the development of these guidelines.

3. Data Collection and Management: Ensure that your data collection and management processes are aligned with privacy and consent regulations. Be mindful of potential biases in the training data and take steps to mitigate them. Strive for a diverse and representative dataset to avoid reinforcing stereotypes or excluding underrepresented groups.

4. Transparency and Explainability: Aim to make your generative AI system transparent and explainable. Users should have a clear understanding of how the system works and what data it uses. Provide explanations for the decisions or outputs generated by the system to build trust and accountability.

5. Bias Detection and Mitigation: Implement methods to detect and mitigate biases in the generated outputs. Regularly audit the system to identify and address any biases that may emerge during training or deployment. Develop mechanisms to allow users to provide feedback on biased outputs and improve the system accordingly.

6. User Control and Consent: Empower users to have control over the generated outputs and ensure they provide informed consent for using their data. Offer options for users to customize or modify the outputs according to their preferences, within ethical boundaries.

7. Ongoing Monitoring and Evaluation: Continuously monitor and evaluate the performance and impact of your generative AI system. Regularly assess its compliance with ethical guidelines and address any issues or concerns that arise. Be responsive to user feedback and iterate on the system to improve its responsible behavior.

8. Collaboration and Accountability: Foster collaboration within the AI community to collectively address the challenges of responsible AI. Engage in discussions, share best practices, and work towards creating industry standards for responsible generative AI systems. Foster a culture of accountability within your organization to ensure responsible deployment and use of AI technologies.
